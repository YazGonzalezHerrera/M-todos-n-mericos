```python
#Spline cbico condicionado
import matplotlib.pyplot as plt
import numpy as np

from scipy.interpolate import CubicSpline
#Damos valores
xs = [0.0,1.0,2.0,3.0]
ys = [1.0, 2.71,7.38,20.08]

#Graficas con respecto a la derivada
clamped_smoother = CubicSpline(xs,ys, bc_type='clamped')

extrapolated = np.linspace(min(xs)-10,max(xs)+10,int(21+(max(xs)-min(xs))))

fig, axs = plt.subplots(3, 3,figsize=(12,12))
axs[0, 0].plot(xs, ys,'.')
axs[0, 0].plot(extrapolated,clamped_smoother(extrapolated),'-')
axs[0, 0].set_title('Spline condicionado, derivada=0')
axs[0, 1].plot(xs, clamped_smoother(xs,nu=1),'.')
axs[0, 1].plot(extrapolated,clamped_smoother(extrapolated,nu=1),'-')
axs[0, 1].set_title('Spline condicionado, derivada=1')
axs[0, 2].plot(xs, clamped_smoother(xs,nu=2),'.')
axs[0, 2].plot(extrapolated,clamped_smoother(extrapolated,nu=2),'-')
axs[0, 2].set_title('Spline condicionado, derivada=2')
plt.show()
```
